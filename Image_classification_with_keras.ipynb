#  Transfer learning

### The goal is to see if we can classify images not originally used to train inception net:
more specifically to find the difference between images of the `houses` and `apartments`

## Load data

```python
from imutils.paths import list_images

N_CLASSES = 2
N_DIM = 224
BATCH_SIZE = 128
```

Now, lets load all the images.

```python
label_names = ['house', 'studio-apartment'] 
path = os.path.expanduser('~/Data/ImageNet/')
paths = list_images(path)
data = []
labels = []
cnt = 0 
for img_path in paths:
    label = img_path.split('/')[-2]
    if label in label_names:
        labels.append(label_names.index(label))
    else:
        continue
    img = image.load_img(img_path, target_size=(N_DIM, N_DIM))
    x = image.img_to_array(img)
    x = preprocess_input(x)
    data.append(x)
    cnt += 1
    
    if cnt % 1e3 == 0:
        print('...', cnt, 'files processed')

print(cnt, 'files read')
data = np.array(data)
labels = np.array(labels)
print(f'data.shape = {data.shape}')
```

###  shuffle data

```python
np.random.seed(11)
idx = np.random.permutation(len(data))
data, labels = data[idx], labels[idx]
```

##  Train on data

```python
from keras.models import Model, Sequential
from keras.layers import Dense, GlobalMaxPooling2D, Dropout
from keras.optimizers import SGD, RMSprop, Adam
```

```python
base_model = InceptionV3(weights='imagenet', include_top=False)
```

```python
# base_model.summary()
```

## transfer learning has two flavors, 
* fine-tuning (where we allow all multiple layers to be tuned, if we have ebought data
* feature extraction (when we only change the last layer of the network)

#### Fine tuning can be done the following way:
    # add a global spatial pooling layer
    outputs = base_model.output
    outputs = GlobalMaxPooling2D()(outputs)
    outputs = Dense(64, activation='relu')(outputs)
    outputs = Dense(N_CLASSES, activation='softmax')(outputs)

    # this is the model we will train
    model = Model(inputs=base_model.input, outputs=outputs)

    for layer in base_model.layers:
        layer.trainable = False
        
    model.compile(optimizer='adam', loss='binary_crossentropy',  metrics=["accuracy"])
    model.fit( ... )
#### To only extract features we run all the images throu the net

```python
features = base_model.predict(data, verbose=1, batch_size=BATCH_SIZE)
```

```python
# globalMaxPooling:
features_pulled = features.max((1,2))

n_labels = np.max(labels) + 1
labels_1hot = np.eye(n_labels, dtype=int)[labels]
class_weights = dict(enumerate(1 / labels_1hot.mean(0)))
input_dim = features.shape[-1]
```

```python
model = Sequential()
model.add(Dense(64,  input_dim=input_dim, kernel_initializer="glorot_uniform", activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(N_CLASSES, activation='softmax'))
model.summary()
optimizer = Adam(lr=0.001)
model.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])
```

Last important bit before training:
> **The classes are not ballanced** 
>
> there are almost twice as many images for houses. So naturally, network will bi biassed towards 'guessing 'house'

to avoid it we downwait houses, like so

```python
class_weights = dict(enumerate(1/labels_1hot.mean(0)))
```

```python
history = model.fit(features_pulled, labels_1hot, epochs=50, validation_split=0.2, shuffle=1, verbose=1,
                    class_weight=class_weights, batch_size=BATCH_SIZE)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
plt.plot(acc)
plt.plot(val_acc)
```
